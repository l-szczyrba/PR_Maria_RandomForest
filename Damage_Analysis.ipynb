```python
import pandas as pd
from pandas.plotting import scatter_matrix

import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
file = "/Users/lszczyrba/Project_3.6.3/PR_Housing/DamageAnalysis.xlsx"
data = pd.read_excel(file)
df = pd.DataFrame(data).set_index('Comm_Code')
```

```python
df
```

```python
DF_statistics = pd.DataFrame(df.groupby('Comm_Code', as_index=True)['R_IND'].mean())
```

```python
DF_statistics['max'] = df.groupby('Comm_Code', as_index=True)['R_IND'].max()
```

```python
DF_statistics
```

```python
df_mean = DF_statistics.sort_values(by=['R_IND'], ascending=False).head(10)
```

```python
df_max = DF_statistics.sort_values(by=['max'], ascending=False).head(10)
```

```python
df_mean
```

```python
df_mean['label'] = ['Ciales', 'Vieques', 'Arroyo', 'Barranquitas', 'Aguas', 'Guayama', 'Vega Alta', 'Naranjito', 'Corozal', 'Catano']
```

```python
df_max
```

```python
df_max['label'] = ['San Juan', 'Toa Alta', 'Bayamon', 'Canovanas', 'Arroyo', 'Vega Baja', 'Catano', 'Arecibo', 'Ciales', 'Toa Baja']
```

```python
df_max
```

```python
df_mean
```

```python
ax = df_mean.sort_values('R_IND', ascending=False).plot.bar('label','R_IND', zorder=3,
                                                                            color='maroon', 
                                                                            legend = False)
plt.title("Municipalities with 10 Highest \n Mean Damage Index Values", fontsize = 25, fontdict={'horizontalalignment': 'center'})
plt.ylabel('Damage Index', fontsize = 18)
plt.xlabel(' ')
plt.xticks(fontsize = 18, rotation = 45, ha='right')
plt.yticks(fontsize = 14)

ax.grid(zorder=0)

#plt.savefig('meandam_munic.png', dpi=600, bbox_inches = "tight")
```

```python
ax = df_max.sort_values('max', ascending=False).plot.bar('label','max',zorder=3,
                                                                            color='maroon', 
                                                                            legend = False)
plt.title("Municipalities with 10 Highest \n Maximum Damage Index Values", fontsize = 25, fontdict={'horizontalalignment': 'center'})
plt.ylabel('Damage Index', fontsize = 18)
plt.xlabel(' ')
plt.xticks(fontsize = 18, rotation = 45, ha='right')
plt.yticks(fontsize = 14)

ax.grid(zorder=0)

#plt.savefig('maxdam_munic.png', dpi=600, bbox_inches = "tight")
```
